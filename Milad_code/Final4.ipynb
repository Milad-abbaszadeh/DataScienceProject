```python
def find_ranking(Team,Year,fifa_ranking):
    temp = fifa_ranking[(fifa_ranking['country_full'] == Team)&(fifa_ranking['rank_date']< Year)].tail(1)
    if temp.shape[0] == 0:
        rank,rank_change,total_points = 0,0,0
    else:
        total_points = temp['total_points'].iloc[0]
        rank_change = temp['rank_change'].iloc[0]
        rank = temp['rank'].iloc[0]
    print(Team, " ",rank,rank_change,total_points)
    return rank,rank_change,total_points    
```
